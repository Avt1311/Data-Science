```python
# 1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?


import scipy.stats as stats
import numpy as np

def margin_of_error(confidence_level, sample_std, sample_size):
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    return z_score * (sample_std / np.sqrt(sample_size))

# Example usage:
confidence_level = 0.95
sample_std = 12
sample_size = 100
moe = margin_of_error(confidence_level, sample_std, sample_size)
print(f"Margin of Error: {moe:.4f}")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?


import numpy as np
import scipy.stats as stats

np.random.seed(42)

sample_size = 50
sample_data = np.random.normal(loc=52, scale=5, size=sample_size)

# Define population parameters
population_mean = 50  # Known population mean
population_std = 5    # Known population standard deviation

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the P-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

# Display the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot?


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
np.random.seed(42)

sample_size = 50
sample_data = np.random.normal(loc=52, scale=5, size=sample_size)

# Define population parameters
population_mean = 50
population_std = 5

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the P-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

# Display the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Visualization
# Define the critical Z-values for a two-tailed test at alpha = 0.05
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)

# Create a range of Z values
z_values = np.linspace(-4, 4, 1000)
z_distribution = stats.norm.pdf(z_values)

# Plot the Z-distribution
plt.figure(figsize=(10, 6))
plt.plot(z_values, z_distribution, label='Standard Normal Distribution', color='blue')

# Fill the critical regions
plt.fill_between(z_values, z_distribution, where=(z_values <= -critical_value) | (z_values >= critical_value),
                 color='red', alpha=0.5, label='Rejection Region')

# Mark the Z-statistic on the plot
plt.axvline(z_statistic, color='green', linestyle='--', label=f'Z-statistic: {z_statistic:.2f}')

# Add labels and title
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

```python
# 5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(population_mean, sample_mean, population_std, sample_size, alpha):
    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Critical Z-value for the given alpha (two-tailed)
    critical_value = stats.norm.ppf(1 - alpha / 2)

    # Calculate Type I Error (Œ±)
    type_1_error = alpha

    # Calculate Type II Error (Œ≤)
    # Assume the true mean is different from the population mean
    true_mean = population_mean + 5  # Example: true mean is 5 units greater
    z_beta = (critical_value * (population_std / np.sqrt(sample_size)) + true_mean - population_mean) / (population_std / np.sqrt(sample_size))
    type_2_error = stats.norm.cdf(z_beta) - stats.norm.cdf(-critical_value)

    # Display the results
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"Type I Error (Œ±): {type_1_error:.4f}")
    print(f"Type II Error (Œ≤): {type_2_error:.4f}")

    # Visualization
    z_values = np.linspace(-4, 4, 1000)
    z_distribution = stats.norm.pdf(z_values)

    plt.figure(figsize=(10, 6))
    plt.plot(z_values, z_distribution, label='Standard Normal Distribution', color='blue')

    # Fill the critical regions for Type I Error
    plt.fill_between(z_values, z_distribution, where=(z_values <= -critical_value) | (z_values >= critical_value),
                     color='red', alpha=0.5, label='Type I Error Region (Œ±)')

    # Fill the Type II Error region
    plt.fill_between(z_values, z_distribution, where=(z_values > -critical_value) & (z_values < z_beta),
                     color='orange', alpha=0.5, label='Type II Error Region (Œ≤)')

    # Mark the Z-statistic on the plot
    plt.axvline(z_statistic, color='green', linestyle='--', label=f'Z-statistic: {z_statistic:.2f}')

    # Add labels and title
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
visualize_errors(population_mean=50, sample_mean=52, population_std=5, sample_size=50, alpha=0.05)
```

```python
# 6 Write a Python program to perform an independent T-test and interpret the results?


import numpy as np
from scipy import stats

# Step 1: Simulate two independent samples
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=100)
group2 = np.random.normal(loc=55, scale=10, size=100)

# Step 2: Perform the independent T-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Step 3: Interpret the results
alpha = 0.05  # Significance level

print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")
```

```python
# 7 Perform a paired sample T-test using Python and visualize the comparison results?


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate related samples
np.random.seed(42)  # For reproducibility
n = 30  # Number of samples
before_treatment = np.random.normal(loc=50, scale=5, size=n)  # Before treatment
after_treatment = before_treatment + np.random.normal(loc=3, scale=2, size=n)  # After treatment (mean increase of 3)

# Step 2: Perform the paired sample T-test
t_statistic, p_value = stats.ttest_rel(before_treatment, after_treatment)

# Step 3: Interpret the results
alpha = 0.05  # Significance level
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")

# Step 4: Visualize the results
plt.figure(figsize=(10, 6))
plt.plot(before_treatment, label='Before Treatment', marker='o', linestyle='-', color='blue')
plt.plot(after_treatment, label='After Treatment', marker='o', linestyle='-', color='orange')
plt.title('Comparison of Before and After Treatment')
plt.xlabel('Sample Index')
plt.ylabel('Measurement')
plt.axhline(y=np.mean(before_treatment), color='blue', linestyle='--', label='Mean Before Treatment')
plt.axhline(y=np.mean(after_treatment), color='orange', linestyle='--', label='Mean After Treatment')
plt.legend()
plt.grid()
plt.show()
```

```python
# 9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (array-like): Sample data.
    confidence (float): Confidence level (default is 0.95 for 95% confidence).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    # Calculate the sample mean and standard error
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)  # Sample size
    standard_error = sample_std / np.sqrt(n)

    # Calculate the critical value for the specified confidence level
    critical_value = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulated sample data
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval for the sample mean: {ci}")
```

```python
# 10 Write a Python program to calculate the margin of error for a given confidence level using sample data?


import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level.

    Parameters:
    data (array-like): Sample data.
    confidence (float): Confidence level (default is 0.95 for 95% confidence).

    Returns:
    float: Margin of error.
    """
    # Calculate the sample standard deviation
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)  # Sample size

    # Calculate the critical value for the specified confidence level
    z_critical = stats.norm.ppf((1 + confidence) / 2)

    # Calculate the margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(n))

    return margin_of_error

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulated sample data
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error for 95% confidence level: {moe:.2f}")
```

```python
# 11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?



import numpy as np

def bayesian_inference(prior, likelihood, evidence):

    posterior = (likelihood * prior) / evidence
    return posterior


prior_disease = 0.01

likelihood_positive_given_disease = 0.9

likelihood_positive_given_no_disease = 0.05

# Marginal likelihood: Total probability of testing positive (P(E))
# P(E) = P(E | H) * P(H) + P(E | ~H) * P(~H)
p_no_disease = 1 - prior_disease  # Probability of not having the disease
marginal_likelihood = (likelihood_positive_given_disease * prior_disease) + \
                      (likelihood_positive_given_no_disease * p_no_disease)

# Calculate the posterior probability (P(H | E))
posterior_probability = bayesian_inference(prior_disease, likelihood_positive_given_disease, marginal_likelihood)

print(f"Posterior Probability of having the disease given a positive test result: {posterior_probability:.4f}")
```

```python
# 12  Perform a Chi-square test for independence between two categorical variables in Python


import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Prepare the data
# Create a contingency table
data = {
    'Variable A': ['Yes', 'Yes', 'No', 'No'],
    'Variable B': ['Group 1', 'Group 2', 'Group 1', 'Group 2'],
    'Count': [30, 10, 20, 40]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Create a contingency table
contingency_table = df.pivot(index='Variable A', columns='Variable B', values='Count')
print("Contingency Table:")
print(contingency_table)

# Step 2: Perform the Chi-square test
chi2_statistic, p_value, dof, expected = chi2_contingency(contingency_table)

# Step 3: Interpret the results
alpha = 0.05  # Significance level
print(f"\nChi-square Statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between the two variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between the two variables.")
```

```python
# 13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data?



import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):

    if not isinstance(observed, pd.DataFrame):
        observed = pd.DataFrame(observed)

    # Calculate the row totals and column totals
    row_totals = observed.sum(axis=1).values.reshape(-1, 1)  # Reshape for broadcasting
    column_totals = observed.sum(axis=0).values  # Column totals

    # Total number of observations
    total = observed.values.sum()

    # Calculate expected frequencies
    expected = (row_totals * column_totals) / total
    return expected

# Example observed data (contingency table)
observed_data = np.array([[30, 10],
                           [20, 40]])

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Display the results
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
 # 14 Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution



 import numpy as np
from scipy.stats import chisquare

# Step 1: Prepare the observed data
observed_data = np.array([50, 30, 20])  # Example observed frequencies

# Step 2: Define the expected distribution
# For example, we expect the proportions to be 50%, 30%, and 20%
total_count = observed_data.sum()
expected_proportions = np.array([0.5, 0.3, 0.2])  # Expected proportions
expected_data = expected_proportions * total_count  # Calculate expected frequencies

# Step 3: Perform the Chi-square goodness-of-fit test
chi2_statistic, p_value = chisquare(observed_data, f_exp=expected_data)

# Step 4: Interpret the results
alpha = 0.05  # Significance level
print(f"Chi-square Statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```

```python
 # 16 Implement an F-test using Python to compare the variances of two random samples



 import numpy as np
from scipy import stats

# Step 1: Simulate two independent samples
np.random.seed(42)  # For reproducibility
sample_size = 30
# Sample 1: Normal distribution with mean=50, std=10
sample1 = np.random.normal(loc=50, scale=10, size=sample_size)
# Sample 2: Normal distribution with mean=55, std=15
sample2 = np.random.normal(loc=55, scale=15, size=sample_size)

# Step 2: Calculate the variances
var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
var2 = np.var(sample2, ddof=1)  # Sample variance for sample2

# Step 3: Perform the F-test
f_statistic = var1 / var2  # F-statistic
df1 = sample_size - 1  # Degrees of freedom for sample1
df2 = sample_size - 1  # Degrees of freedom for sample2

# Calculate the p-value
p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

# Step 4: Interpret the results
alpha = 0.05  # Significance level
print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

```python
# 17 D Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results


import numpy as np
from scipy import stats

# Step 1: Simulate data for multiple groups
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=5, size=30)  # Group 1: mean=50, std=5
group2 = np.random.normal(loc=55, scale=5, size=30)  # Group 2: mean=55, std=5
group3 = np.random.normal(loc=60, scale=5, size=30)  # Group 3: mean=60, std=5

# Step 2: Perform the ANOVA test
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Step 3: Interpret the results
alpha = 0.05  # Significance level
print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between group means.")
```

```python
# 19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def check_anova_assumptions(data):

    # Check normality using Shapiro-Wilk test
    print("Normality Test Results:")
    for column in data.columns:
        stat, p_value = stats.shapiro(data[column])
        print(f"{column}: W-statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"  -> {column} is not normally distributed (reject H0)\n")
        else:
            print(f"  -> {column} is normally distributed (fail to reject H0)\n")

    # Check equal variance using Levene's test
    print("Equal Variance Test Results:")
    stat, p_value = stats.levene(*[data[column] for column in data.columns])
    print(f"Levene's test: W-statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("  -> Variances are not equal (reject H0)\n")
    else:
        print("  -> Variances are equal (fail to reject H0)\n")

    # Visualize the data distribution
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=data)
    plt.title('Boxplot of Groups')
    plt.ylabel('Values')
    plt.xlabel('Groups')
    plt.grid()
    plt.show()

# Example usage
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=55, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)

# Combine the data into a DataFrame
data = pd.DataFrame({
    'Group 1': group1,
    'Group 2': group2,
    'Group 3': group3
})

# Check ANOVA assumptions
check_anova_assumptions(data)
```

```python
# 20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Simulate data for two factors
np.random.seed(42)  # For reproducibility
factor1 = ['A', 'B']  # Factor 1 with two levels
factor2 = ['X', 'Y', 'Z']  # Factor 2 with three levels

# Create a DataFrame to hold the data
data = []
for f1 in factor1:
    for f2 in factor2:
        # Simulate data: mean varies by factor levels
        mean = 10 + (ord(f1) - ord('A')) * 5 + (ord(f2) - ord('X')) * 3
        group_data = np.random.normal(loc=mean, scale=2, size=30)
        data.extend(zip([f1]*30, [f2]*30, group_data))

df = pd.DataFrame(data, columns=['Factor1', 'Factor2', 'Response'])

# Step 2: Perform the Two-Way ANOVA test
model = ols('Response ~ C(Factor1) * C(Factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)

# Step 3: Visualize the results
# Interaction plot
plt.figure(figsize=(10, 6))
sns.pointplot(x='Factor1', y='Response', hue='Factor2', data=df,
               palette='deep', markers=['o', 's', 'D'], dodge=True)
plt.title('Interaction Plot for Two-Way ANOVA')
plt.ylabel('Response')
plt.xlabel('Factor 1')
plt.grid()
plt.show()
```

```python
# 23 Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means


import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=5, size=30)  # Group 1: mean=50, std=5
group2 = np.random.normal(loc=55, scale=5, size=30)  # Group 2: mean=55, std=5

# Combine the data into a DataFrame for easier plotting
data = pd.DataFrame({
    'Group 1': group1,
    'Group 2': group2
})

# Step 2: Formulate Hypotheses
# Null Hypothesis (H0): The means of the two groups are equal (Œº1 = Œº2)
# Alternative Hypothesis (H1): The means of the two groups are not equal (Œº1 ‚â† Œº2)

# Step 3: Perform the T-test
t_statistic, p_value = stats.ttest_ind(data['Group 1'], data['Group 2'])

# Step 4: Interpret the results
alpha = 0.05  # Significance level
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the means.")

# Step 5: Visualize the results
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Box Plot of Group Means')
plt.ylabel('Values')
plt.xlabel('Groups')
plt.grid()
plt.axhline(y=np.mean(group1), color='blue', linestyle='--', label='Mean Group 1')
plt.axhline(y=np.mean(group2), color='orange', linestyle='--', label='Mean Group 2')
plt.legend()
plt.show()
```

```python
# 24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results


import numpy as np
import scipy.stats as stats

# Step 1: Formulate Hypotheses
# Null Hypothesis (H0): The population variance is equal to a specified value (œÉ¬≤ = œÉ0¬≤)
# Alternative Hypothesis (H1): The population variance is not equal to the specified value (œÉ¬≤ ‚â† œÉ0¬≤)

# Step 2: Simulate data
np.random.seed(42)  # For reproducibility
sample_size = 30
population_variance = 25  # Assume the true population variance is 25
sample_data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)

# Step 3: Calculate the Chi-square Statistic
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (Bessel's correction)
chi_square_statistic = (sample_size - 1) * sample_variance / population_variance

# Step 4: Determine the Critical Value
alpha = 0.05  # Significance level
df = sample_size - 1  # Degrees of freedom
critical_value_lower = stats.chi2.ppf(alpha / 2, df)  # Lower critical value
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)  # Upper critical value

# Step 5: Interpret the Results
print(f"Chi-square Statistic: {chi_square_statistic:.2f}")
print(f"Critical Value Lower: {critical_value_lower:.2f}")
print(f"Critical Value Upper: {critical_value_upper:.2f}")

if chi_square_statistic < critical_value_lower or chi_square_statistic > critical_value_upper:
    print("Reject the null hypothesis: The population variance is significantly different from the specified value.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from the specified population variance.")
```

```python
# 25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups?

import numpy as np
import scipy.stats as stats

# Step 1: Define the data
# Group 1: successes and total trials
successes1 = 50  # Number of successes in group 1
trials1 = 200    # Total trials in group 1

# Group 2: successes and total trials
successes2 = 30  # Number of successes in group 2
trials2 = 150    # Total trials in group 2

# Step 2: Formulate Hypotheses
# Null Hypothesis (H0): p1 = p2 (the proportions are equal)
# Alternative Hypothesis (H1): p1 ‚â† p2 (the proportions are not equal)

# Step 3: Calculate the Z-statistic
# Calculate the proportions
p1 = successes1 / trials1
p2 = successes2 / trials2

# Calculate the pooled proportion
p_pool = (successes1 + successes2) / (trials1 + trials2)

# Calculate the standard error
standard_error = np.sqrt(p_pool * (1 - p_pool) * (1/trials1 + 1/trials2))

# Calculate the Z-statistic
z_statistic = (p1 - p2) / standard_error

# Step 4: Determine the p-value
# For a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Step 5: Interpret the Results
alpha = 0.05  # Significance level
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the proportions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions.")
```

```python
# 26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate two independent samples
np.random.seed(42)  # For reproducibility
sample_size = 30
# Sample 1: Normal distribution with mean=50, std=10
sample1 = np.random.normal(loc=50, scale=10, size=sample_size)
# Sample 2: Normal distribution with mean=55, std=15
sample2 = np.random.normal(loc=55, scale=15, size=sample_size)

# Step 2: Perform the F-test
# Calculate the variances
var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
var2 = np.var(sample2, ddof=1)  # Sample variance for sample2

# Calculate the F-statistic
f_statistic = var1 / var2  # F-statistic
df1 = sample_size - 1  # Degrees of freedom for sample1
df2 = sample_size - 1  # Degrees of freedom for sample2

# Calculate the p-value
p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

# Step 3: Interpret the results
alpha = 0.05  # Significance level
print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Step 4: Visualize the results
plt.figure(figsize=(10, 6))
plt.boxplot([sample1, sample2], labels=['Sample 1', 'Sample 2'])
plt.title('Box Plot of Two Samples')
plt.ylabel('Values')
plt.grid()
plt.axhline(y=np.mean(sample1), color='blue', linestyle='--', label='Mean Sample 1')
plt.axhline(y=np.mean(sample2), color='orange', linestyle='--', label='Mean Sample 2')
plt.legend()
plt.show()
```

```python
# 27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results?



import numpy as np
import pandas as pd
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Step 1: Simulate observed data
# Simulate observed frequencies for a categorical variable (e.g., colors of balls)
np.random.seed(42)  # For reproducibility
observed_data = np.random.choice(['Red', 'Blue', 'Green'], size=100, p=[0.5, 0.3, 0.2])
observed_counts = pd.Series(observed_data).value_counts()

# Step 2: Define expected frequencies
# Assume we expect the proportions to be 50% Red, 30% Blue, and 20% Green
total_count = observed_counts.sum()
expected_counts = np.array([0.5 * total_count, 0.3 * total_count, 0.2 * total_count])

# Step 3: Perform the Chi-square test
chi2_statistic, p_value = chisquare(observed_counts, f_exp=expected_counts)

# Step 4: Analyze the results
alpha = 0.05  # Significance level
print(f"Observed Counts:\n{observed_counts}")
print(f"Expected Counts:\n{expected_counts}")
print(f"Chi-square Statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

# Visualization
plt.figure(figsize=(10, 6))
plt.bar(observed_counts.index, observed_counts.values, alpha=0.6, label='Observed', color='blue', width=0.4)
plt.bar(observed_counts.index, expected_counts, alpha=0.6, label='Expected', color='orange', width=0.4)
plt.title('Observed vs Expected Counts')
plt.ylabel('Counts')
plt.xlabel('Categories')
plt.legend()
plt.show()
```
